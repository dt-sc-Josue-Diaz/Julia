```julia
@gif for x ∈ 0:0.1:5π
    plot(0:0.01:5π, sin.(0:0.01:5π),cos.(0:0.01:5π) , key = false, color = :green)
    scatter!([x+i for i in 0:0.3:10], [sin(x+i) for i in 0:0.3:10], [cos(x+i) for i in 0:0.3:10], show = :ijulia, color = :pink, 
        alpha = 0.2, markerstrokewidth = 0.5, markersize = 10) 
end
```

# Bases de datos

```julia
using CSV
using DataFrames
```

```julia
# Crear un data frame
Calif = DataFrame(Pepe=[], Diana=[], Tono=[])
```

```julia
# Agreagar contenido, push!()
push!(Calif.Pepe,10)
```

```julia
# Indicar datos nulos
push!(Calif.Diana, missing)
push!(Calif.Tono,missing)
```

```julia
Calif
```

```julia
# Cambiar el tipo de dato del dataframe
Calif = DataFrame(Pepe=Float64[], Diana=Float64[], Tono=Float64[])
```

```julia
# Sin embargo "missing" no es un tipo flotante. Ahora es exclusiva de float. 
allowmissing!(Calif)
```

```julia
push!(Calif.Pepe,10)
push!(Calif.Diana, missing)
push!(Calif.Tono,missing)
```

```julia
Calif
```

```julia
# leer bases de datos
calif = DataFrame(CSV.File("calif.csv"))
```
